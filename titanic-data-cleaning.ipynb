```python
import pandas as pd 

data = pd.read_csv('/kaggle/input/datasets/yasserh/titanic-dataset/Titanic-Dataset.csv') # for load csv data .
df=data.copy() # make the copy of original data because if i lost teh data i will get it back .
df.head() # for understand , what kind of data i have 
```

```python
# data cleaning
 
df.info() # for get the all info about data 
df.isnull().sum()
```

# Clean the data 

1. Column names are inconsistent , so i convert them into lower case and the space will fill by _ .
2. IN this data 714 cell of age column are null , so i use (.groupby) in pclass to get the median of every pclass then fill the age cell by the     median of thier pclass.
3. In cabin column 77% of data is null so i am  delete the cabin column,  because fill the null cells with 0 or average is not making sense.
4. Embarked means here "The port where passenger boarded the titanic , embarks column has only two cell of missing data and teh "s" 644 passenger are boarded from "s" so i fill the missing cell with "s".
5. The dataset has no duplicate value.

# 

```python
df.columns=(df.columns.str.lower().str.strip().str.replace(" ","_"))
df["age"]= df["age"].fillna(df.groupby("pclass")["age"].transform("median"))
df.drop("cabin",axis=1,inplace=True)
df["embarked"]=df["embarked"].fillna(df["embarked"].mode()[0])
df.duplicated().sum()
```
